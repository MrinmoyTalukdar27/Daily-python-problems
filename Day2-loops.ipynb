```python
# a program to print a table 
Num=int(input('Enter the number'))
i=1
while i<11:
    print(Num* i)
    i +=1
```

```python
#while loops with else
x=1
while x<3:
    print(x)
    x+=1
else:
    print("limit crossed")
```

```python
# A gussing game
#first we need to generate a random number using random module
import random
jacpot= random.randint(1,100)

guess= int(input('Guss karo'))
counter=1
#here we are adding and while loop and the loop will keep running untill jacpot number = user number
while guess !=jacpot:
#Here we are telling the user if his guss is higher or lower using if-else statement
    if guess<jacpot:
        print(' Galat guess higher')
    else:
        print('Galat guess lower')
# Here we are came out from the if-else statement but not from the loop and now we are give the user another chance to guss
# we will keep giving them chance to guesss untill he gets it
    guess=int(input('Again guss karo'))
    counter+=1

# once he guss the number we will came out of that loop and printthe below statement
# and we will also tell the user about his atteps like how many times he gussed using counter
else:
    print('correct guess')
    print('Attempts',counter)
```
