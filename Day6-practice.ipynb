```python
# Write a program to separate words in a string using a given delimiter (space'').without using split function ()
M=input('Enter your string')
L=[]# we took an empty list
D=' '#given delimiter which is space

for i in M:
    if i!=' ':
        D=D+i
    else:
        L.append(D)
        D=''
L.append(D)
print('your string length is=',L)
```

```python
# write a python program to convert a string to title case without using title case
M=input('Enter the string')
L=[]
for i in M.split():
    L.append(i[0].upper()+i[1:].lower())
print(' '.join(L))

```
