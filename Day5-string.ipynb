Straing

```python
# positive indexing 
# use to access straing char from left to right
s="Hello"
print(s[2])

# Nagative Indexing
# use to access strings char from last element
M="world"
print(M[-3])
```

```python
# Slicing in straing
M= "Hello World"
print(M[6:11])

# when we skip the last element it print the straing from its given index here its 0 till the end of the straing
# same if we skip the first it will frint from the first index till the given index here its 4
M2="Hi Friend"
print(M2[0:])
print(M2[:4])

# we can also pass steps (its mean how many steps we nee need to jump/skip.)
M3="who_are_you"
print(M3[0:6:2])# here step size is 2 so its going to jump 2nd numbers

# to reverse a string
print(M3[::-1])
```

```python
# In a given string Hello_world print the world and reverse it
M="Hello_world"
print(M[-1:-6:-1])
```
