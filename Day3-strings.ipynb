### Today i will do some left over loop probles and after that we will do string problems 

```python
# Now we need to solve a problem where 1/1!+2/2!+3/3!....+n
# The n term will be given by the user
N=int(input("Enter the N number"))
result=0
fact=1

for i in range(1,N+1):
    fact=fact*i
    result=result+i/fact
print(result)
```

#### Nested loops problems

```python
# we need to print the pairs like 11,12,13,14,21,22...
for i in range(1,3):
    for j in range(1,3):
        print(i,j)
```

```python
# Make this pattrern
#*
#**
#***
#****
rows=int(input("number of rows"))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print('*',end="")
    print()
```

```python
# solve this pattrern
#1
#121
#12321
#1234321
num=int(input("Enetr the row number"))
for i in range(1,num+1):
    for j in range(1,i+1):
        print(j,end="")
    for k in range(i-1,0,-1):
        print(k,end="")

    print()
```

```python
# user will give us a range and we need to print all the prime numbers between that range
lower=int(input("give the lower range"))
high=int(input("Enter the higher number"))
for i in range(lower,high+1):
    for j in range(2,i):
        if i%j ==0:
            break
    else:
        print(i) 
```
